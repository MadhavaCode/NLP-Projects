Monica is tasked with analyzing customer reviews using sentiment analysis. Given the following reviews:

"This product is fantastic! I'm so happy with my purchase."

"Terrible experience. I want a refund!"

"The quality is okay, but not worth the price."

Instructions:

What sentiment (positive, negative, or neutral) would the sentiment analyzer assign to each of these reviews?
How confident is the model in its analysis? Explain the scores in terms of the sentiment.


HINT:

Library Import: The program uses the pipeline function from the transformers library, which provides access to pre-trained models for various NLP tasks, including sentiment analysis.

Sentiment Analysis Pipeline Initialization: The pipeline("sentiment-analysis") loads a pre-trained sentiment analysis model to classify text as positive or negative with a confidence score.

Sentiment Analysis Function:

It defines a function that takes a list of texts and a category name as input.
It analyzes each text's sentiment using the pre-trained model.
For each text, it prints the sentiment label (e.g., Positive or Negative) and the confidence score (a value between 0 and 1).
Customer Review Challenge:

A specific function is created to analyze customer reviews.
A sample set of customer reviews is provided, including positive, negative, and neutral feedback.
The function calls the sentiment analysis function to classify the sentiment of each review.
Program Execution:

The main block of the script ensures the program runs only when executed directly.
It prints a welcome message and then calls the function to analyze customer reviews.


```python
import transformers
from transformers import pipeline
```

```python
sentimentAnalyzer = pipeline("sentiment-analysis")
```

```python
reviews = [
    "This product is fantastic! I'm so happy with my purchase.",
    "Terrible experience. I want a refund!",
    "The quality is okay, but not worth the price.",
    "Loving the new update! #Awesome",
    "I can't believe how bad this service has gotten...",
    "Does anyone else think this app is overrated?",
    "The instructor was very clear and helpful.",
    "The class felt rushed and lacked structure.",
    "Great content, but more examples would be helpful.",
    "I found the book very inspiring and relatable.",
    "The topic was confusing, but the discussion helped.",
    "I didn't agree with the arguments presented."
]
```

```python
for review in reviews:
    result = sentimentAnalyzer(review)[0]
    print(f"Review: {review}")
    print(f"Sentiment: {result['label']}")
    print(f"Confidence Score: {result['score']}")
```
