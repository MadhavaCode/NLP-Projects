```python
from googletrans import Translator
```

```python
def translate_to_japanese(english_text):
  translator = Translator()
  translation = translator.translate(english_text, dest="ja")
  return translation.text
```

```python
english_phrase = "Thank you very much"
japanese_translation = translate_to_japanese(english_phrase)
print(f'Translation of "{english_phrase}" into Japanese is: "{japanese_translation}"')
```

```python
def FrenchToEnglish(FrenchText):
  translator = Translator()
  translation = translator.translate(FrenchText, dest="en")
  return translation.text
```

```python
french_phrase = "merci beaucoup"
english_translation = FrenchToEnglish(french_phrase)
print(f'Translation of "{french_phrase}" into English is: "{english_translation}"')
```

```python
def translate_to_german(english_text):
  translator = Translator()
  translation = translator.translate(english_text, dest="de")
  return translation.text
```

```python
english_phrase = "The world is beautiful."
german_translation = translate_to_german(english_phrase)
print(f'Translation of "{english_phrase}" into German is: "{german_translation}"')
```
